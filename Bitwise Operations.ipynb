```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img1=cv2.imread('1bit1.png')
img2=cv2.imread('2bit2.png')
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
# bitwise and
cv2.imshow('Image 1 ',img1)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('Image 2 ',img2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
dest_and=cv2.bitwise_and(img1,img2,mask=None)
```

```python
plt.imshow(dest_and)
```

```python
cv2.imshow('And image',dest_and)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Bitwise OR
dest_or=cv2.bitwise_or(img1,img2,mask=None)
```

```python
plt.imshow(dest_or)
```

```python
cv2.imshow('OR image',dest_or)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#NOT operation
dest_not=cv2.bitwise_not(img1,mask=None)
```

```python
plt.imshow(dest_not)
```

```python
cv2.imshow('NOT image',dest_not)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
